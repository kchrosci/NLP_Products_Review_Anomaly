```python
import pandas as pd
from googletrans import Translator

# Mapowanie flag jƒôzyka na kody jƒôzyka obs≈Çugiwane przez translate
lang_flags = {
    'GBR': 'en',
    'DEU': 'de',
}

# Wczytaj plik CSV
df = pd.read_csv('csv_data/nlp2023_v3.csv')

# Utw√≥rz obiekt t≈Çumacza
translator = Translator()

# Iteruj po wierszach i t≈Çumacz opisy
rows_to_drop = []  # Lista indeks√≥w wierszy do usuniƒôcia

for index, row in df.iterrows():
    lang = row['language']
    text = row['content']
    
    if lang not in ['POL']:

        try:
            lang_code = lang_flags.get(lang)
        except Exception as e:
            print(f"Error while mapping language code for {lang}: {str(e)}")
        
        if lang not in lang_flags:
            lang_code = translator.detect(text).lang

        if lang_code:
            try:
                translated_text = translator.translate(text, dest='pl', src=lang_code).text
                df.at[index, 'content'] = translated_text
                df.at[index, 'language'] = lang

            except:
                rows_to_drop.append(index)
                continue
    
    if index % 50 == 0:
        print(index)

# Usu≈Ñ nieprzet≈Çumaczone wiersze
df = df.drop(rows_to_drop)

# Zapisz wynik do nowego pliku CSV
df.to_csv('csv_data/nlp2023_v3_translated.csv', index=False)
```
