```python
# Wczytaj pierwszy plik CSV
df1 = pd.read_csv('csv_data/nlp2023_v2.csv')

# Wczytaj drugi plik CSV
df2 = pd.read_csv('csv_data/doubleQuality.csv')

# Wybierz interesujące kolumny z pierwszego pliku
columns1 = ['salusBcContent.description', 'salusBcContent.language', 'salusBcContent.doubleQuality']
df1_columns = df1[columns1].copy()

# Wybierz interesujące kolumny z drugiego pliku
columns2 = ['content', 'language', 'doubleQuality']
df2_columns = df2[columns2].copy()

# Zmień nazwy kolumn z pierwszego pliku na Text, Origin i IsDoubleQuality
df1_columns.rename(columns={'salusBcContent.description': 'Text', 'salusBcContent.language': 'Origin', 'salusBcContent.doubleQuality': 'IsDoubleQuality'}, inplace=True)

# Zmień nazwy kolumn z drugiego pliku na Text, Origin i IsDoubleQuality
df2_columns.rename(columns={'content': 'Text', 'language': 'Origin', 'doubleQuality': 'IsDoubleQuality'}, inplace=True)

# Zamień wartości IsDoubleQuality na 1 i 0 w pierwszym pliku
df1_columns['IsDoubleQuality'] = df1_columns['IsDoubleQuality'].map({True: 1, False: 0})

# Zamień wartości IsDoubleQuality na 1 i 0 w drugim pliku
df2_columns['IsDoubleQuality'] = df2_columns['IsDoubleQuality'].map({True: 1, False: 0})

# Połącz dane z obu plików
merged_df = pd.concat([df1_columns, df2_columns], ignore_index=True)

# Zapisz wynik do nowego pliku CSV
merged_df.to_csv('csv_data/nlp2023_merged.csv', index=False)
```

```python
import pandas as pd
from googletrans import Translator

# Mapowanie flag języka na kody języka obsługiwane przez translate
lang_flags = {
    'GBR': 'en',
    'DEU': 'de',
}

# Wczytaj plik CSV
df = pd.read_csv('csv_data/doubleQuality_V2.csv')

# Utwórz obiekt tłumacza
translator = Translator()

# Iteruj po wierszach i tłumacz opisy
rows_to_drop = []  # Lista indeksów wierszy do usunięcia

for index, row in df.iterrows():
    lang = row['Origin']
    text = row['Text']
    
    if lang not in ['POL']:

        try:
            lang_code = lang_flags.get(lang)
        except Exception as e:
            print(f"Error while mapping language code for {lang}: {str(e)}")
        
        if lang not in lang_flags:
            lang_code = translator.detect(text).lang

        if lang_code:
            try:
                translated_text = translator.translate(text, dest='pl', src=lang_code).text
                print(translated_text)
                df.at[index, 'Text'] = translated_text
                df.at[index, 'Origin'] = lang

            except TypeError as e:
                print(f"Error while translating text: {str(e)}")
                rows_to_drop.append(index)
                continue

    print(index)

# Usuń nieprzetłumaczone wiersze
df = df.drop(rows_to_drop)

# Zapisz wynik do nowego pliku CSV
df.to_csv('csv_data/doubleQuality_final.csv', index=False)
```
